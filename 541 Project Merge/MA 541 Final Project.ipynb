# Fitting ETF Data to a Cauchy and Normal Distribution

```python
fit_parameters_cauchy_etf = stats.cauchy.fit(data['Close_ETF'])
fit_parameters_norm_etf = stats.norm.fit(data['Close_ETF'])

x = np.arange(0,200,1)
pdf_cauchy_etf = stats.cauchy.pdf(x,fit_parameters_cauchy_etf[0],fit_parameters_cauchy_etf[1])
pdf_norm_etf = stats.cauchy.pdf(x,fit_parameters_norm_etf[0],fit_parameters_norm_etf[1])

plt.plot(x,pdf_cauchy_etf,color = 'red')
plt.hist(data['Close_ETF'], bins = 25)
plt.title('ETF Cauchy Fit')
plt.xlabel('Daily Change')
plt.ylabel('Count Frequency')
plt.show()

cauchy_test_etf = stats.kstest(data['Close_ETF'],"cauchy",fit_parameters_cauchy_etf)
norm_test_etf = stats.kstest(data['Close_ETF'],"norm",fit_parameters_norm_etf)

print('Cauchy Test Results:',cauchy_test_etf)
```

```python
plt.plot(x,pdf_norm_etf,color = 'red')
plt.hist(data['Close_ETF'],bins=25)
plt.title('ETF Normal Fit')
plt.xlabel('Daily Change')
plt.ylabel('Count Frequency')
plt.show()

print('Normal Test Results:',cauchy_test_etf)
```

```python
ETF_mean = data['Close_ETF'].mean()
ETF_std = data['Close_ETF'].std()
print('ETF mean =',ETF_mean,'\nETF std = ',ETF_std)
```

# I.
We cut the ETF Data into 50 subsets of size 20 and calculate their respective means

```python
data['group_1'] = pd.cut(data['Close_ETF'],bins = 50, labels = False)

group_means = []
for i in range(50):
    mu = data[data['group_1'] == i]['Close_ETF'].mean()
    group_means.append(mu)
```

```python
subsets = pd.DataFrame({'Mean':group_means})
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
subsets['Mean'].hist(bins = 50)
plt.xlabel('Subset Mean')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of ETF Subset (size 20) Means')
```

The histogram is not normal and skewed, and does not appear to follow any well known distribution. 

```python
mean_of_subsets_20 = subsets['Mean'].mean()
std_of_subsets_20 = subsets['Mean'].std()
std_over_root_50 = ETF_std/np.sqrt(50) 

print('Mean of subsets =',mean_of_subsets_20,'\n','std of subsets =',std_of_subsets_20)
print('\n')
print('ETF Pupulation mean =',ETF_mean) 
print('ETF Population std =',ETF_std)
print('\n')
print('n=50, ','(ETF Population std)/root(n) = ',std_over_root_50)
```

The sample mean is larger than the mean of sequential subsets of size 20. 

The standard deviation of the subset means is larger than the sample standard deviation over root(n), where n = 50.

The results above are not consistent with the Central Limit Theorem, because the sample of subsets was made sequentially, not randomly.
