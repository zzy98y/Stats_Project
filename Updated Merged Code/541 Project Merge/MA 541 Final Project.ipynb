# Fitting ETF Data to a Cauchy and Normal Distribution

```python
fit_parameters_cauchy_etf = stats.cauchy.fit(data['Close_ETF'])
fit_parameters_norm_etf = stats.norm.fit(data['Close_ETF'])

x = np.arange(0,200,1)
pdf_cauchy_etf = stats.cauchy.pdf(x,fit_parameters_cauchy_etf[0],fit_parameters_cauchy_etf[1])
pdf_norm_etf = stats.cauchy.pdf(x,fit_parameters_norm_etf[0],fit_parameters_norm_etf[1])

fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
plt.plot(x,pdf_cauchy_etf,color = 'red')
plt.hist(data['Close_ETF'], bins = 25)
plt.title('ETF Closing Price Cauchy Fit')
plt.xlabel('Closing Price')
plt.ylabel('Count Frequency')
plt.show()

cauchy_test_etf = stats.kstest(data['Close_ETF'],"cauchy",fit_parameters_cauchy_etf)
norm_test_etf = stats.kstest(data['Close_ETF'],"norm",fit_parameters_norm_etf)

print('Cauchy Test Results:',cauchy_test_etf)
```

The p-value is significantly less than .05, therefore we reject the null hypothesis that the data fits a Cauchy Distribution. Moreover, the plot clearly does not show any resemblance. 

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
plt.plot(x,pdf_norm_etf,color = 'red')
plt.hist(data['Close_ETF'],bins=25)
plt.title('ETF Closing Price Normal Fit')
plt.xlabel('Closing Price')
plt.ylabel('Count Frequency')
plt.show()

print('Normal Test Results:',cauchy_test_etf)
```

The p-value is significantly less than .05, therefore we reject the null hypothesis that the data fits a Normal Distribution. Moreover, the plot clearly does not show any resemblance.

```python
ETF_mean = data['Close_ETF'].mean()
ETF_std = data['Close_ETF'].std()
print('ETF mean =',ETF_mean,'\nETF std = ',ETF_std)
```

# I.
We sequentially cut the ETF Data into 50 subsets of size 20 and calculate their respective means. Next, given the resulting distribution of the sequential subset means, we calculate its mean and standard deviation.

```python
data['group_1'] = pd.cut(data['Close_ETF'],bins = 50, labels = False)

group_means = []
for i in range(50):
    mu = data[data['group_1'] == i]['Close_ETF'].mean()
    group_means.append(mu)
```

```python
subsets = pd.DataFrame({'Mean':group_means})
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
subsets['Mean'].hist(bins = 50)
plt.xlabel('Subset Mean')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of ETF Subset (size 20) Means')
plt.show()
```

The histogram is not normal and skewed. It does not appear to follow any well known distribution. 

```python
mean_of_subset_means_20 = subsets['Mean'].mean()
std_of_subset_means_20 = subsets['Mean'].std()
std_over_root_50 = ETF_std/np.sqrt(50) 

print('Mean of subset means =',mean_of_subsets_20,'\n','std of subset means =',std_of_subsets_20)
print('\n')
print('ETF Pupulation mean =',ETF_mean) 
print('ETF Population std =',ETF_std)
print('\n')
print('n=50, ','(ETF Population std)/root(n) = ',std_over_root_50)
```

The average of the sequential subset means is larger than the ETF population mean by about 3. 

The standard deviation of the sequential subset means is much larger than the ETF population standard deviation over root(n), where n = 50. 

The results above are not consistent with the Central Limit Theorem because the sample of subsets was made sequentially, not randomly. Moreover, the number of subsets created is nowhere near large enough at only 50.

We may be able to get closer to a Normal Distribution if we take n=100 random samples of size 10. 

A related question to consider is if number of values drawn overall is allowed to exceed 1000. If not, we would have more freedom to make n as large as necessary in order to see if a Normal distribution arises.
